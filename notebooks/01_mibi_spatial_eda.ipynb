```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
# 1. Load the 16-bit or multi-channel TIFF file
# Replace with the actual path to one of your downloaded MIBI images
image_path = r"C:\Users\Kelly\Documents\spatial-ihc-feature-lab\data\raw\p31_labeledcellData.tiff"
img = io.imread(image_path)
```

```python
# 2. Check the shape of the data
print(f"Image shape: {img.shape}")
print(f"Max pixel value: {img.max()}")
```

```python
# 3. Choose a single channel to view (if it's a multi-channel stack)
# If the image shape is (Channels, Height, Width), select channel 0
if len(img.shape) > 2:
    channel_to_view = img[0, :, :] 
else:
    channel_to_view = img
```

```python
# 4. Stretch the contrast to ignore extreme outliers and make the signal visible
vmax_adjusted = np.percentile(channel_to_view, 99) # Set max brightness to the 99th percentile

plt.figure(figsize=(8, 8))
plt.imshow(channel_to_view, cmap='magma', vmin=0, vmax=vmax_adjusted)
plt.title(f"MIBI Image (Contrast Stretched to max {vmax_adjusted:.1f})")
plt.axis('off')
plt.show()
```
